```python
import os
import glob
import pandas as pd

data_dir = "./datasets/data/DIOR/coco"
pickle_files = glob.glob(os.path.join(data_dir, "*.pkl"))

for file_path in pickle_files:
    print("DataFrame:", file_path)
    df = pd.read_pickle(file_path)
    
    print("Original file_path:")
    print(df['file_path'].head(2))
    
    df['file_path'] = df['file_path'].str.replace('/home/bibahaduri/DIOR/coco/', '')
    df.to_pickle(file_path)
    
    print("Modified file_path:")
    print(df['file_path'].head(2))
```
