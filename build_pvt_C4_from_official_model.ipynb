```python
import torch

model = torch.load(model_name)
```

```python
# Define a fixed width for the layer names
layer_name_width = 40

for layer_ in list(model.keys()):
    glob_match = layer_.split(".")[0][-1] # Extract the number at the end of the prefix to decide backbone or roi_heads
    if glob_match in ["1", "2", "3"]:
        new_key = 'backbone.' + layer_
    elif glob_match == "4":
        new_key = 'roi_heads.' + layer_
    else:
        continue
    model[new_key] = model.pop(layer_)
    # Pad the layer name with spaces to the fixed width
    padded_layer_name = layer_.ljust(layer_name_width)
    print(f"replacing {padded_layer_name} with {new_key}")
```

```python
torch.save(model, model_name.replace(".pth", "_C4.pth"))
model.keys()
```
